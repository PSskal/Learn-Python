```python
from tkinter import *
from random import randint
from datetime import datetime
import csv

root = Tk()

root.title("Menu Cash Register")
root.resizable(0,0)
#variable globales
Precios = []
Cant = []

#Obj-variables:
var_num1 = IntVar()
var_num2 = IntVar()
var_num3 = IntVar()
var_num4 = IntVar()
var_num5 = IntVar()
var_num6 = IntVar()

var_refer = StringVar()
var_costo = StringVar()
var_delivery = StringVar()
var_impuesto = StringVar()
var_total = StringVar()
var_Pfinal = StringVar()

var_Prom = BooleanVar()
var_Nombre = StringVar()

#Funciones:
def dicionario(a):
    combos = {}
    with open(a,encoding='latin-1') as csv_file:  
        reader = csv.reader(csv_file,delimiter=";")
        next(reader)
        for data_line in reader:
            data_line[1] = float(data_line[1])
            combos[data_line[0]] = data_line[1]
        
        lista_precios = []
        for nomb,precio in combos.items():
            lista_precios.append(precio)
            
    return lista_precios

Precios = dicionario("precios.csv")

def total(lista1):
    cantidad = [var_num1.get(), var_num2.get(), var_num3.get(), var_num4.get(), var_num5.get(), var_num6.get()]
    t = []
    for i in range(len(lista1)):
        t.append(lista1[i]*cantidad[i])
    
    costo = sum(t)
    delivery = costo*0.03
    impuesto = costo*0.18
    Total = costo+delivery+impuesto
    if var_Prom.get() == True:
        Pfinal = Total*0.9
    else:
        Pfinal = Total
    
    var_refer.set(randint(10000,99999))
    var_costo.set("{:.2f}".format(costo))
    var_delivery.set("{:.2f}".format(delivery))
    var_impuesto.set("{:.2f}".format(impuesto))
    var_total.set("{:.2f}".format(Total))
    var_Pfinal.set("{:.2f}".format(Pfinal))
    
    btnRecibo.config(state='normal')
    
def Eliminar():
    var_refer.set('')
    var_costo.set('')
    var_delivery.set('')
    var_impuesto.set('')
    var_total.set('')
    var_Pfinal.set('')
    
    var_num1.set(0)
    var_num2.set(0)
    var_num3.set(0)
    var_num4.set(0)
    var_num5.set(0)
    var_num6.set(0)
    
    CbtnProm.deselect()
    entNombre.delete(0, END)
    btnRecibo.config(state='disable')

def generaRecibo():
    with open("Recibo.txt",mode="w",encoding="utf-8") as file:
        file.write("BURGER MAX\n")
        file.write("==========\n")
        file.write("Ref. Pedido: {}\n".format(var_refer.get()))
        file.write(f"Fecha: {datetime.now():%d/%m/%Y}\n")
        file.write(f"Hora: {datetime.now():%H:%M:%S}\n")
        file.write("Nombre: {}\n\n".format(var_Nombre.get()))
        file.write("rarezza:\n")
        file.write("==============================\n\n")
        file.write("Costo: {} PEN\n".format(var_costo.get()))
        file.write("Delivery [3%]: {} PEN\n".format(var_delivery.get()))
        file.write("Impuesto [18%]: {} PEN\n".format(var_impuesto.get()))
        file.write("Total: {} PEN\n\n".format(var_total.get()))
        file.write("==============================\n\n")
        file.write("Precio Final: {} PEN\n".format(var_Pfinal.get()))
    
    btnRecibo.config(state='disable')

#Frames:
frm = Frame(root)
frmUp = Frame(frm)
frmDown = Frame(frm)

frm.pack(padx=10,pady=10)
frmUp.pack(padx=10,pady=10)
frmDown.pack(padx=10,pady=10)

#Agregando widgets al frmUp:
lblCombo1 = Label(frmUp,text="Combo Clásico")
lblCombo2 = Label(frmUp,text="Combo Doble Queso")
lblCombo3 = Label(frmUp,text="Combo Tocino")
lblCombo4 = Label(frmUp,text="Combo Royal")
lblCombo5 = Label(frmUp,text="Combo Parrillero")
lblCombo6 = Label(frmUp,text="Combo Light")

lblReferencia = Label(frmUp,text="Referencia")
lblCosto = Label(frmUp,text="Costo")
lblDelivery = Label(frmUp,text="Delivery")
lblImpuesto = Label(frmUp,text="Impuesto")
lblTotal = Label(frmUp,text="Total")
lblPFinal = Label(frmUp,text="Precio Final")

entCombo1 = Entry(frmUp,width=10,textvariable=var_num1)
entCombo2 = Entry(frmUp,width=10,textvariable=var_num2)
entCombo3 = Entry(frmUp,width=10,textvariable=var_num3)
entCombo4 = Entry(frmUp,width=10,textvariable=var_num4)
entCombo5 = Entry(frmUp,width=10,textvariable=var_num5)
entCombo6 = Entry(frmUp,width=10,textvariable=var_num6)

entReferencia = Entry(frmUp,width=20,state='disable',textvariable=var_refer)
entCosto = Entry(frmUp,width=20,state='disable',textvariable=var_costo)
entDelivery = Entry(frmUp,width=20,state='disable',textvariable=var_delivery)
entImpuesto = Entry(frmUp,width=20,state='disable',textvariable=var_impuesto)
entTotal = Entry(frmUp,width=20,state='disable',textvariable=var_total)
entPFinal = Entry(frmUp,width=20,state='disable',textvariable=var_Pfinal)

lblCombo1.grid(row=0,column=0,padx=5,pady=5,sticky=E)
lblCombo2.grid(row=1,column=0,padx=5,pady=5,sticky=E)
lblCombo3.grid(row=2,column=0,padx=5,pady=5,sticky=E)
lblCombo4.grid(row=3,column=0,padx=5,pady=5,sticky=E)
lblCombo5.grid(row=4,column=0,padx=5,pady=5,sticky=E)
lblCombo6.grid(row=5,column=0,padx=5,pady=5,sticky=E)

lblReferencia.grid(row=0,column=3,padx=5,pady=5,sticky=E)
lblCosto.grid(row=1,column=3,padx=5,pady=5,sticky=E)
lblDelivery.grid(row=2,column=3,padx=5,pady=5,sticky=E)
lblImpuesto.grid(row=3,column=3,padx=5,pady=5,sticky=E)
lblTotal.grid(row=4,column=3,padx=5,pady=5,sticky=E)
lblPFinal.grid(row=5,column=3,padx=5,pady=5,sticky=E)

entCombo1.grid(row=0,column=1,padx=5,pady=5)
entCombo2.grid(row=1,column=1,padx=5,pady=5)
entCombo3.grid(row=2,column=1,padx=5,pady=5)
entCombo4.grid(row=3,column=1,padx=5,pady=5)
entCombo5.grid(row=4,column=1,padx=5,pady=5)
entCombo6.grid(row=5,column=1,padx=5,pady=5)

entReferencia.grid(row=0,column=4,padx=5,pady=5)
entCosto.grid(row=1,column=4,padx=5,pady=5)
entDelivery.grid(row=2,column=4,padx=5,pady=5)
entImpuesto.grid(row=3,column=4,padx=5,pady=5)
entTotal.grid(row=4,column=4,padx=5,pady=5)
entPFinal.grid(row=5,column=4,padx=5,pady=5)

#Agregando widgets al frmDown:
lblNombre = Label(frmDown,text="Nombre cliente")
entNombre = Entry(frmDown,width=25,textvariable=var_Nombre)
CbtnProm = Checkbutton(frmDown,text="Aplicar promoción?",variable=var_Prom)

btnTotal = Button(frmDown, text="Total",width=20, command=lambda: total(Precios))
btnReset = Button(frmDown, text="Reset",width=20, command=lambda: Eliminar())
btnRecibo = Button(frmDown, text="Recibo",width=20,state='disable', command=lambda:generaRecibo())

lblNombre.grid(row=0,column=0,padx=5,pady=5,sticky=E)
entNombre.grid(row=0,column=1,padx=5,pady=5)
CbtnProm.grid(row=0,column=2,padx=5,pady=5)

btnTotal.grid(row=1,column=0,padx=5,pady=5)
btnReset.grid(row=1,column=1,padx=5,pady=5)
btnRecibo.grid(row=1,column=2,padx=5,pady=5)

root.mainloop()
```
