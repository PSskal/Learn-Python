```python
#Richard Pascual Huallpa Yapo --- U20201C125
#jhean Brando Sánchez Del Aguila ---U20201B927
#Wilmer Efrain Garcia Argumedo --- U20201B406

from tkinter import *
from random import randint
from datetime import datetime


root = Tk()

root.title("Menu Cash Register")
root.resizable(0,0)


#objeto variables
var_Cclasic = IntVar()
var_Cdqueso = IntVar()
var_Ctocino = IntVar()
var_Croyal = IntVar()
var_Cparri = IntVar()
var_Clight = IntVar()

var_Nclient = StringVar()
var_Apromo = BooleanVar()

var_Ref = IntVar()
var_Costo = IntVar()
var_Delivery = IntVar()
var_Impuesto = IntVar()
var_Total = IntVar()
var_PrFinal = IntVar()

#abrir archivo
with open("precios.csv", encoding="utf-8") as archivo:
    next(archivo)  # Omitir encabezado del archivo
    precios = {}
    for linea in archivo:
        linea = linea.rstrip("\n")  # Quitar salto de línea
        columnas = linea.split(";")
        nombre = columnas[0]
        costox1 = float(columnas[1])
        precios[nombre]=costox1
    
#funcion para generar aleatorio
numAlea=0
def aleatorio():
    """
    * se genera una un numero aletorio de de 5 digitos
    * se retorna el numero aleatorio
    """
    global numAlea
    numAlea=randint(10000,100000)
    
    return numAlea


def probando(): #nada inmportando solo prbando con el boton reset
    global final
    result = var_Cclasic.get()*precios["combo1"]+var_Cdqueso.get()*precios["combo2"]+var_Ctocino.get()*precios["combo3"]+var_Croyal.get()*precios["combo4"]+var_Cparri.get()*precios["combo5"]+var_Clight.get()*precios["combo6"]
    total=round(result+0.03*result+0.18*result, 2)
    if var_Apromo.get() == 1:
        final =total-(total*0.10)
        final=round(final,2)
        return final
    else:
        return total
        
 #funciones 
def activarbotton():
    
    """activando el boton recibo"""
    
    if (btnRecibo['state'] == DISABLED):
        btnRecibo['state'] = NORMAL

    
    global result 
    global delivery
    global impuesto
    global total
    result = var_Cclasic.get()*precios["combo1"]+var_Cdqueso.get()*precios["combo2"]+var_Ctocino.get()*precios["combo3"]+var_Croyal.get()*precios["combo4"]+var_Cparri.get()*precios["combo5"]+var_Clight.get()*precios["combo6"]
    
    delivery=result*0.03
    impuesto=result*0.18
    total=result+0.03*result+0.18*result
    
    var_Ref.set(aleatorio())
    var_Costo.set("{:.2f}".format(result)) 
    var_Delivery.set("{:.2f}".format(delivery))
    var_Impuesto.set("{:.2f}".format(impuesto))
    var_Total.set("{:.2f}".format(total))
    var_PrFinal.set(probando())    
    
def cambiarbotton():
    """Desactivando el boton recibo y borrar los campos"""
  
    if (btnRecibo['state'] == NORMAL):
        btnRecibo['state'] = DISABLED
        

    entNclient.delete(0,END)
    var_Apromo.set(0)
        
    var_Cclasic.set(0)
    var_Cdqueso.set(0)
    var_Ctocino.set(0)
    var_Croyal.set(0)
    var_Cparri.set(0)
    var_Clight.set(0)
          
    var_Ref.set("")
    var_Costo.set("") 
    var_Delivery.set("")
    var_Impuesto.set("")
    var_Total.set("")
    var_PrFinal.set("") 
    

        
def generarArchivo():
    
    if (btnRecibo['state'] == NORMAL):
        btnRecibo['state'] = DISABLED
        
    with open("recibo_"+str(numAlea)+".txt",mode="w") as file:
    
        file.write("BURGER MAX\n")
        file.write("="*10+"\n")
        file.write("Ref.Pedido: "+str(numAlea)+"\n")
        file.write(f"Fecha: {datetime.now():%d/%m/%Y}\n")
        file.write(f"Hora: {datetime.now():%H:%M:%S}\n")
        file.write("Nombre: "+str(var_Nclient.get())+"\n\n")
        if var_Cclasic.get() > 0:
            file.write(f'Combo clásico:{precios["combo1"]:>12.2f} PEN\n')
        if var_Cdqueso.get() > 0:
            file.write(f'Combo doble queso:   {precios["combo2"]:.2f} PEN\n')
        if var_Ctocino.get() > 0 :
            file.write(f'Combo tocino:        {precios["combo3"]:.2f} PEN\n')
        if var_Croyal.get() > 0:
            file.write(f'Combo royal:         {precios["combo4"]:.2f} PEN\n')
        if var_Cparri.get() > 0:
            file.write(f'Combo parrillero:    {precios["combo5"]:.2f} PEN\n')
        if var_Clight.get() > 0:
            file.write(f'Combo light:         {precios["combo6"]:.2f} PEN\n')
        file.write("="*30+"\n\n\n")
        file.write(f"Costo:               {result:.2f} PEN\n")
        file.write(f"Delivery [3%]:        {delivery:.2f} PEN\n")
        file.write(f"Impuesto [18%]:       {impuesto:.2f} PEN\n")
        file.write(f"Total:               {total:.2f} PEN\n\n")
        file.write("="*30+"\n\n")
        if var_Apromo.get() == 1:
            file.write(f"Precio Final:       {final:.2f} PEN")
        elif var_Apromo.get() == 0:
            file.write(f"Precio Final:       {total:.2f} PEN")
        
        
        


frmPrincipal = Frame(root)
frmSecundario = Frame(frmPrincipal)
frmSecundario1 = Frame(frmPrincipal)

frm1 = Frame(frmSecundario)
frm2 = Frame(frmSecundario)
frm3 = Frame(frmSecundario1)


frmPrincipal.pack(padx=10,pady=10)
frmSecundario.pack(padx=10,pady=0)
frmSecundario1.pack(side=LEFT,padx=35,pady=0)

frm1.pack(anchor=N,side=LEFT,padx=10,pady=10)
frm2.pack(anchor=N,side=RIGHT,padx=10,pady=10)
frm3.pack(anchor=S,padx=10,pady=10)
#frmPrincipal.pack(padx=10,pady=10)

##FRAME 1
lblcclasic = Label(frm1,text="Combo clasico")
lblcdqueso = Label(frm1,text="Combo doble queso")
lblctocino = Label(frm1,text="Combo tocino")
lblcroyal = Label(frm1,text="Combo royal")
lblcparri = Label(frm1,text="Combo parrillero")
lblclight = Label(frm1,text="Combo light")

entcclasic = Entry(frm1,width=13,textvariable=var_Cclasic)
entcdqueso = Entry(frm1,width=13,textvariable=var_Cdqueso)
entctocino = Entry(frm1,width=13,textvariable=var_Ctocino)
entcroyal = Entry(frm1,width=13,textvariable=var_Croyal)
entcparri = Entry(frm1,width=13,textvariable=var_Cparri)
entclight = Entry(frm1,width=13,textvariable=var_Clight)

#GRID 1
lblcclasic.grid(row=0,column=0,padx=5,pady=5,sticky=E)
lblcdqueso.grid(row=1,column=0,padx=5,pady=5,sticky=E)
lblctocino.grid(row=2,column=0,padx=5,pady=5,sticky=E)
lblcroyal.grid(row=3,column=0,padx=5,pady=5,sticky=E)
lblcparri.grid(row=4,column=0,padx=5,pady=5,sticky=E)
lblclight.grid(row=5,column=0,padx=5,pady=5,sticky=E)

entcclasic.grid(row=0,column=1,padx=5,pady=5)
entcdqueso.grid(row=1,column=1,padx=5,pady=5)
entctocino.grid(row=2,column=1,padx=5,pady=5)
entcroyal.grid(row=3,column=1,padx=5,pady=5)
entcparri.grid(row=4,column=1,padx=5,pady=5)
entclight.grid(row=5,column=1,padx=5,pady=5)



#FRAME 2
lblref = Label(frm2,text="Referencia")
lblcost = Label(frm2,text="Costo")
lbldeliv = Label(frm2,text="Deliveri")
lblimpuest = Label(frm2,text="Impuesto")
lbltotal = Label(frm2,text="Total")
lblpfinal = Label(frm2,text="Precio final")

entref = Entry(frm2,width=23,state='disabled', textvariable=var_Ref)
entcost = Entry(frm2,width=23,state='disabled',textvariable=var_Costo)
entdeliv = Entry(frm2,width=23,state='disabled',textvariable=var_Delivery)
entimpuest = Entry(frm2,width=23,state='disabled',textvariable=var_Impuesto)
enttotal = Entry(frm2,width=23,state='disabled',textvariable=var_Total)
entpfinal = Entry(frm2,width=23,state=DISABLED,textvariable=var_PrFinal)


#GRID 2
lblref.grid(row=0,column=0,padx=5,pady=5,sticky=E)
lblcost.grid(row=1,column=0,padx=5,pady=5,sticky=E)
lbldeliv.grid(row=2,column=0,padx=5,pady=5,sticky=E)
lblimpuest.grid(row=3,column=0,padx=5,pady=5,sticky=E)
lbltotal.grid(row=4,column=0,padx=5,pady=5,sticky=E)
lblpfinal.grid(row=5,column=0,padx=5,pady=5,sticky=E)

entref.grid(row=0,column=1,padx=5,pady=5)
entcost.grid(row=1,column=1,padx=5,pady=5)
entdeliv.grid(row=2,column=1,padx=5,pady=5)
entimpuest.grid(row=3,column=1,padx=5,pady=5)
enttotal.grid(row=4,column=1,padx=5,pady=5)
entpfinal.grid(row=5,column=1,padx=5,pady=5)

#frame 3
lblNclient = Label(frm3,text="Nombre cliente",)
entNclient = Entry(frm3,width=30,textvariable=var_Nclient)

chkApromo = Checkbutton(frm3,text="Aplicar promocion?", variable=var_Apromo,command=probando)

btnTotal = Button(frm3,text="Total",width=12, command = activarbotton) #areglar
btnReset = Button(frm3,text="Reset",width=12, command = cambiarbotton) #corregirt
btnRecibo = Button(frm3,text="Recibo",width=12,state=DISABLED, command = generarArchivo)

#grid
lblNclient.grid(row=0,column=0,padx=5,pady=5,sticky=E)
entNclient.grid(row=0,column=1,padx=5,pady=5)
chkApromo.grid(row=0,column=2,padx=5,pady=5)

btnTotal.grid(row=1,column=0,padx=5,pady=5)
btnReset.grid(row=1,column=1,padx=5,pady=5)
btnRecibo.grid(row=1,column=2,padx=5,pady=5)

root.mainloop()
```
