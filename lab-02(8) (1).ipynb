Dado el código de la celda de abajo, complete las instrucciones del bloque *with* de tal manera que se tabule la función: 
           \begin{equation}
               f(x)=\frac{cos(x).x^3}{x^2 + 1}
           \end{equation}

Para el dominio de *x* perteneciente al intervalo [-7 , 7] (utilice 100 elementos para el dominio).

**Obs.:** Utilice las herramientas de *numpy* vistas en clase para la creación de los arreglos de dominio y rango.

Todos los valores de *x* y los respectivos valores de la tabulación deberán escribirse en el archivo *tabulación.csv* (en formato excel), respetando el

                           x     F(x)
                           -     -
                           -     -
                           -     -
                           -     -
                           .     .
                           .     .
                           .     .
                           .     .  
                           
Finalmente, se deberá graficar la función *F(x)* vs *x*, usando el siguente formato:
          
          - línea segmentada de color rojo
          - cuadrícula en la gráfica
          - título: "Función matemática"
          - etiqueta (eje x): "dominio [x]"
          - etiqueta (eje y): "rango [y]"
          - marcadores tipo cuadrado, con color de fondo azúl.
          
**Obs.:** Utilice las herramientas de *matplotlib* vistas en clase para la generación de la gráfica

```python
import numpy as np
import matplotlib.pyplot as plt
import csv


y = (np.cos(x)*x**3)/(x**2+1)
x = np.linspace(-7,7,100)

data = dict(zip(x,y))

with open("tabulacion.csv","w",newline="") as csv_file:
    
    writer= csv.writer(csv_file,delimiter=";")  
    writer.writerow(["x","f(x)"])
    for k,v in data.items():
        writer.writerow([k,v])

fig = plt.figure(figsize=(10,10))
plt.plot(x,y,"--rs")
plt.title("Función Matemática")
plt.xlabel("dominio [x]")
plt.ylabel("rango [y]")
plt.grid()
plt.show()
    
```

### Problema 5 (2 ptos.) : 

Dado el código de la celda de abajo, complete las instrucciones dentro del *for* de la función *creaConjuntos(n)* de tal manera que retorne una lista de *n* conjuntos, donde cada conjunto tendrá 5 números aleatorios generados en el intervalo [10,50]

Por ejemplo, si la función se invoca como *print(creaConjuntos(3))*, la lista retornada podría ser: 
 
    [{44, 15, 50, 23, 26}, {43, 13, 16, 24, 31}, {40, 11, 16, 50, 25}]

```python
from random import randint

def creaConjuntos(n):
    num = 5
    lista = []
    l = []
    c = set()
    for i in range(1,n+1):
        while len(c) < num:
            lista.append(randint(10,50))
            c = set(lista)
        l.append(c)
            
    return l

print(creaConjuntos(3))
```

### Problema 7 (4 ptos.) : 

Complete el código de la celda de abajo de tal manera que permita construir la clase *Nota*, la cual definirá a cualquier objeto instanciado a través del atributo público *valor*. La clase deberá controlar que al atributo *valor* solo se le asigne valor numérico (int o float) y en el rango [0,20], de lo contrario se genererán las excepciones de tipo o valor correspondiente. La impresión (representación) de cualquier objeto deberá respetar el formato "Nota[valor:x]".

La clase deberá definir la operación obj1%obj2 (donde obj1 y obj2 representan a objetos de la clase a construir) que dará como resultado otro objeto de la misma clase cuyo atributo será igual a:  $(valor1 + valor2)/2$ (*valor1* y *valor2* son los atributos de obj1 y obj2)

**Obs.:** En python cuando se ejecuta la operación % entre 2 objetos, se llama implícitamente al método mágico **__mod__**

```python
class Nota:
    
    def __init__(self, valor=0):
        self.valor=valor
        
    def __repr__(self):
        return "Nota[valor:{}]".format(self.valor)        
    @property
    def valor(self):
        return self.__valor
    
    @valor.setter
    def valor(self,val):
        if isinstance(val,int) or isinstance(val,float):
            if val>0 and val<=20: #in range(0,20):
                self.__valor=val
            else:
                raise ValueError("el rango es [0,20]")
        else:
            raise TypeError("debe ser int or float")
            
    def __mod__(self,other):
        resul=(self.valor+other.valor)/2
        return "Nota[valor:{}]".format(resul)
```

```python
#comprobación de la clase:
obj1 = Nota(valor=10.5)
obj2 = Nota(valor=3)

print(obj1,obj2) #deberia salir: Nota[valor:10.5] Nota[valor:3]

obj1.valor = 12.5
obj2.valor = 16
print(obj1.valor, obj2.valor)  #debería salir: 12.5 16

obj3 = obj1%obj2
print(obj3) #debería salir: Nota[valor:14.25]

#obj1.valor = -4  #debe salir excepción de valor
#obj2.valor = "UPC"  #debe salir excepción de tipo
```
