```python
#graficando una función matemática:

#creando el arreglo de dominio (abcisa):
x = np.linspace(-10,10,1000)

#creando el arreglo de rango (ordenada):
y = np.sin(np.pi*x)/(np.pi*x)

#graficando y vs x:
plt.plot(x,y)
plt.title("Función sinc")
plt.xlabel("Dominio [x]")
plt.ylabel("Rango [y]")
plt.grid()
plt.show()
```

```python
#agregando mas formato a la gráfica:

#creando el arreglo de dominio (abcisa):
x = np.linspace(-10,10,1000)

#creando el arreglo de rango (ordenada):
y = np.sin(np.pi*x)/(np.pi*x)

#graficando y vs x:
plt.plot(x,y,color='r',linewidth=3)
plt.title("Función sinc")
plt.xlabel("Dominio [x]")
plt.ylabel("Rango [y]")
plt.grid()
plt.show()
```

```python
#aplicación:

class Circulo:  #versión-01
    
    def __init__(self,radio=1,centro=(0,0)):
        self.radio=radio
        self.centro=centro
        
    def __repr__(self):
        return "Circulo[radio={:.2f},centro={}]".format(self.radio,self.centro)
    
    def area(self):
        return np.pi*self.radio**2
    
    def graficar(self,ancho):  
        teta = np.linspace(0,2*np.pi,100)
        x = self.radio*np.cos(teta) + self.centro[0]
        y = self.radio*np.sin(teta) + self.centro[1]
        plt.plot(x,y)
        plt.axis("square")
        plt.axis([-ancho/2,ancho/2,-ancho/2,ancho/2])
        
    def moverArriba(self,paso):
        self.centro = (self.centro[0],self.centro[1]+paso)
        
    def moverAbajo(self,paso):
        self.centro = (self.centro[0],self.centro[1]-paso) 
        
    def moverIzquierda(self,paso):
        self.centro = (self.centro[0]-paso,self.centro[1])
        
    def moverDerecha(self,paso):
        self.centro = (self.centro[0]+paso,self.centro[1])
```

```python
c1 = Circulo(radio=5,centro=(2,4))
print(c1)

fig = plt.figure(figsize=(8,8))

print("Area:",c1.area())
c1.graficar(20)

c1.moverDerecha(3)
print(c1)
fig.clf()
c1.graficar(20)
plt.show()
```

```python
#creando una lista de circulos, por comprehensión:

circulos = [Circulo(radio=i) for i in range(1,11)]

fig = plt.figure(figsize=(8,8))

for c in circulos:
    print(c)
    
for c in circulos:
    c.graficar(25)
    
plt.show()
```

```python
#aplicación:

class Circulo:  #versión-02
    
    def __init__(self,radio=1,centro=(0,0)):
        self.radio=radio
        self.centro=centro
        
    def __repr__(self):
        return "Circulo[radio={:.2f},centro={}]".format(self.radio,self.centro)
    
    def area(self):
        return np.pi*self.radio**2
    
    def generaEjes(self):  
        teta = np.linspace(0,2*np.pi,100)
        x = self.radio*np.cos(teta) + self.centro[0]
        y = self.radio*np.sin(teta) + self.centro[1]
        return x,y
    
    @property
    def radio(self):
        return self.__radio
    
    @radio.setter
    def radio(self,val):
        if isinstance(val,int) or isinstance(val,float):
            if val>0:
                self.__radio=val
            else:
                raise ValueError("El radio debe ser positivo")
        else:
            raise TypeError("El radio debe ser int o float")
            
            
    @property
    def centro(self):
        return self.__centro
    
    @centro.setter
    def centro(self,val):
        if isinstance(val,tuple):
            if len(val)==2:
                aux1 = isinstance(val[0],int) or isinstance(val[0],float)
                aux2 = isinstance(val[1],int) or isinstance(val[1],float)
                if aux1 and aux2:
                    self.__centro=val
                else:
                    raise TypeError("Los 2 valores del centro deben ser numéricos")
            else:
                raise  ValueError("La tupla debe tener 2 elementos")
        else:
            raise TypeError("EL centro debe ser una tupla")
```

```python
#comprobando el control del radio:
c = Circulo()
c.radio = 5.6
print(c)
```

```python
#comprobando el control del centro:
c = Circulo()
c.centro = (5.6,6)
print(c)
```

```python
#creando una lista de circulos, por comprehensión:(con legenda)

circulos = [Circulo(radio=i) for i in range(1,6)]

fig = plt.figure(figsize=(8,8))

for c in circulos:
    x,y = c.generaEjes()
    plt.plot(x,y,label="Area:{:.2f}".format(c.area()))
    
plt.legend(loc = 'lower right')
plt.show()
```

```python
#ejercicio aplicativo:

numDados = int(input("Número de dados a lanzar: "))
numVeces = int(input("Número de lanzamientos: "))

lanzamientos = np.random.randint(1,7,(numDados,numVeces))
#print(lanzamientos)

asuma = np.sum(lanzamientos,axis=0)
#print(asuma)

minimo = numDados
maximo = 6*numDados

x = np.arange(minimo,maximo+1)
#print(x)

y = np.array([])  #arreglo vacio
for v in x:
    veces = asuma[asuma==v].size
    y = np.append(y,veces/numVeces)
    
#print(y)
plt.bar(x,y)
plt.show()
```

```python
#dividir una figura en subventanas: uso de subplot()

fig = plt.figure(figsize=(8,8))

x = np.arange(0,10,0.25)
y1 = x**2
y2 = 2*x**2
y3 = 3*x**2
y4 = 4*x**2

plt.subplot(2,2,1)
plt.plot(x,y1,'--b')

plt.subplot(2,2,2)
plt.plot(x,y2,':ro')

plt.subplot(2,2,3)
plt.plot(x,y3,'-gs')

plt.subplot(2,2,4)
plt.plot(x,y4,':y*')

plt.tight_layout()

plt.show()
```
